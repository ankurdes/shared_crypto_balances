```python
full_node = HttpProvider('https://api.trongrid.io')
solidity_node = HttpProvider('https://api.trongrid.io')
event_server = HttpProvider('https://api.trongrid.io')
tron = Tron(full_node=full_node,
            solidity_node=solidity_node,
            event_server=event_server)

headers = {
    'Content-Type': "application/json",
    'TRON-PRO-API-KEY': TRON_API_KEY
    }

TRON_API_URL = 'https://api.trongrid.io/wallet/triggersmartcontract'
```

```python
async def main():
  cg = CoinGeckoAPI()
  total_bsc_wallet_value = 0
  total_eth_wallet_value = 0
  total_tron_wallet_value = 0

  def get_cg_price(token):
    return cg.get_price(ids=token, vs_currencies='usd')[token]['usd']

  def get_float(str_val, token_dec):
    _ = int(str_val) // 10**(int(token_dec)-2)
    return _ / 100

  print(f"----------------------------")
  print(f"----------  BSC  -----------")
  print(f"----------------------------")
  async with BscScan(BSC_API_KEY) as bsc:
    for token in BSC_TOKENS:
      token_add, token_dec = BSC_TOKENS[token].split('|')
      bal = await bsc.get_acc_balance_by_token_contract_address(contract_address=token_add,address=WALLET)
      token_balance = get_float(bal, token_dec)
      token_price = get_cg_price(token)
      token_usd_value = token_balance * token_price
      total_bsc_wallet_value = total_bsc_wallet_value + token_usd_value
      print(f"{token}: ${token_usd_value} --- {token_balance} tokens with price of ${token_price}")
      time.sleep(0.2)
    print(f"----------------------------")
    print(f"TOTAL USD VALUE: ${total_bsc_wallet_value}")
    print(f"----------------------------")

  print(f"")
  print(f"")
  print(f"----------------------------")
  print(f"----------  ETH  -----------")
  print(f"----------------------------")
  infura_url = 'https://mainnet.infura.io/v3/1061861275ad40329f78fd0cc115ac4d'
  ethereum_address = WALLET
  ethbalance = pyetherbalance.PyEtherBalance(infura_url)
  for token in ETH_TOKENS:
    token_add, token_dec = ETH_TOKENS[token].split('|')
    details = {'symbol': token, 'address': token_add, 'decimals': int(token_dec), 'name': token}
    erc20tokens = ethbalance.add_token(token, details)
    token_balance = ethbalance.get_token_balance(token, WALLET)['balance']
    token_price = get_cg_price(token)
    token_usd_value = token_balance * token_price
    total_eth_wallet_value = total_eth_wallet_value + token_usd_value
    print(f"{token}: ${token_usd_value} --- {token_balance} tokens with price of ${token_price}")
  print(f"----------------------------")
  print(f"TOTAL USD VALUE: ${total_eth_wallet_value}")
  print(f"----------------------------")

  print(f"")
  print(f"")
  print(f"----------------------------")
  print(f"----------  TRON  -----------")
  print(f"----------------------------")
  for token in TRON_TOKENS:
    token_add = TRON_TOKENS[token]
    contract_address = tron.address.to_hex(token_add)
    owner_address = tron.address.to_hex(TRON_WALLET)
    parameter = '0000000000000000000000' + owner_address
    payload = "{\n    \"contract_address\": \"" + contract_address + "\",\n    \"function_selector\": \"balanceOf(address)\",\n    \"parameter\": \"" + parameter + "\",\n    \"owner_address\": \"" + owner_address + "\"\n}"
    response = requests.request("POST", TRON_API_URL, data=payload, headers=headers)
    token_balance = tron.toInt(hexstr=json.loads(response.text)['constant_result'][0]) / 1000000
    token_price = get_cg_price(token)
    token_usd_value = token_balance * token_price
    total_tron_wallet_value = total_tron_wallet_value + token_usd_value
    print(f"{token}: ${token_usd_value} --- {token_balance} tokens with price of ${token_price}")
  print(f"----------------------------")
  print(f"TOTAL USD VALUE: ${total_tron_wallet_value}")
  print(f"----------------------------")
```
